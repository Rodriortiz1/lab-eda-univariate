```python
#Part 2: Delving into Product Pricing
```

```python
#1 Calculate the mean, median, and mode for the price of products.
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]

mean_price, median_price, mode_price
```

```python
#2 What's the average price point of products listed? How does this compare with the most common price point (mode)?
#podriamos decir que a pesar de que el precio de producto que mas se repite es 9.99, la media es de 89,24. Estoq quiere decir que hay productos con precios muy elevados que hacen que la media sea muy elevada con respecto a la moda. 
```

```python
#3 Determine the variance, standard deviation, range, and interquartile range for product price.
variance_price = df['price'].var()
std_dev_price = df['price'].std()
min_price = df['price'].min()
max_price = df['price'].max()
range_price = max_price - min_price
quantiles_price = df['price'].quantile([0.25, 0.5, 0.75])

variance_price, std_dev_price, min_price, max_price, range_price, quantiles_price
```

```python
#4 How varied are the product prices? Are there any indicators of a significant spread in prices?
# Podemos ver que la variacion de precios en los productos es muy alta debido al valor elevado que tiene el rango 
```

```python
#Is there a specific price range where most products fall? Plot a histogram to visualize the distribution of product prices. If its hard to read these diagrams, think why this is, and explain how it could be solved..
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=False, bins=20, color="salmon")
plt.title("Distribución de Precios de Productos")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()
```

```python
#Are there products that are priced significantly higher than the rest? Use a box plot to showcase the spread and potential outliers in product pricing.
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data= df["price"], color = "lightblue") 
plt.title("diagrama de caja de precios de productos")
plt.show
```
